```python
def show_ims(data):
    cmap=plt.get_cmap('Blues')
    fig = plt.figure(figsize=(10,15))
    fig.patch.set_facecolor('white')
    im=data[0]
    ax=[]
    for i in range(6):
        ax.append(plt.subplot(3, 2, i+1))
        plt.imshow(im[i],cmap=cmap,extent=[-0.2,0.2,-0.2,0.2])
        plt.title(ds.layer_names[i])
        plt.xlabel("$\Delta\phi$")
        plt.ylabel("$\Delta\eta$")
        plt.colorbar()
```

```python
from torch.utils.data import DataLoader
from torch.utils.data.sampler import SubsetRandomSampler
train_iter=DataLoader(ds,batch_size=64,shuffle=False,sampler=SubsetRandomSampler(ds.train_indices),num_workers=8)
val_iter=DataLoader(ds,batch_size=64,shuffle=False,sampler=SubsetRandomSampler(ds.val_indices),num_workers=8)
test_iter=DataLoader(ds,batch_size=64,shuffle=False,sampler=SubsetRandomSampler(ds.test_indices),num_workers=8)
```

```python
def loop_over_set(loader):

    # Let's measure time that takes in each loop
    trecord = np.zeros(len(loader),dtype=np.float32)
    t = time.time()
    for iteration, batch in enumerate(loader):

        data, labels = batch
        if iteration%100==0:
            print('iteration {}'.format(iteration))
            print('data: {}', data)
            print('labels: {}', labels)
        

        # Print out some content info
        #print('Iteration',iteration,'... time:',time.time()-t,'[s]')
        #print('    Labels:',labels)

        trecord[iteration] = time.time() - t
        t = time.time()

        # break when reaching the loop limit
        
    return trecord
```

```python
%%time
t_arr=loop_over_set(train_iter)
```
